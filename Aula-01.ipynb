```python
def comparar(a,b):
  return a>b
```

```python
def trocar(lista,a,b):
  lista[a],lista[b] = lista[b],lista[a]

if comparar(lista[0],lista[1]):
  trocar(lista,0,1)
print(lista)
```

```python
lista.reverse()
print(lista)
```

```python
print(lista)

inicio = time.time()

tam = len(lista)

for i in range(tam-1):
  for j in range(tam-1-i):
    if comparar(lista[j],lista[j+1]):
      trocar(lista,j,j+1)

fim = time.time()
tempo = fim - inicio
print(tempo)
print(lista)
```
