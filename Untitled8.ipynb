```python
listarandint = [random.randint(1,100) for i in range(10)]

print(listarandint)

listarandint.sort(reverse=True)

print(listarandint)
```

```python
for i in range(len(listarandint)):
  menor = i
  for j in range(i+1,len(listarandint)):
    if listarandint[j]<listarandint[menor]:
      menor = j
    listarandint[i],listarandint[menor] = listarandint[menor],listarandint[i]
print(listarandint)
```

```python
listarandint = [random.randint(1,100) for i in range(10)]

print(listarandint)

for i in range(len(listarandint)):
  chave = i
  temp = listarandint[chave]
  for j in range(chave-1,-1,-1):
    if temp < listarandint[j]:
      listarandint[chave] = listarandint[j]
      chave -=1
      listarandint[chave] = temp

print(listarandint)
```
